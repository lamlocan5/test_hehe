I. EXTRACT FROM OXFORD

```python
file_path = r"E:\Project\2025_Project\English_audio_extractor_from_online_dict\unit2.csv"
df = pd.read_csv(file_path)

output_dir = r"E:\Project\2025_Project\English_audio_extractor_from_online_dict\audio"
os.makedirs(output_dir, exist_ok=True)

# Th√™m headers gi·∫£ l·∫≠p tr√¨nh duy·ªát
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Referer": "https://www.oxfordlearnersdictionaries.com/",
    "Accept-Language": "en-US,en;q=0.9",
}

# D√πng Session ƒë·ªÉ gi·ªØ k·∫øt n·ªëi
session = requests.Session()
session.headers.update(HEADERS)

def download_audio(word, dem):
    url = f"https://www.oxfordlearnersdictionaries.com/definition/english/{word}"
    
    try:
        response = requests.get(url, headers=HEADERS, timeout=10)  # Gi·ªõi h·∫°n timeout 10s
        if response.status_code != 200:
            print(f"‚ùå Kh√¥ng t√¨m th·∫•y t·ª´: {word} (HTTP {response.status_code})")
            return False

        # Ph√¢n t√≠ch HTML
        soup = BeautifulSoup(response.text, "html.parser")
        audio_tag = soup.find("div", class_="sound audio_play_button pron-uk icon-audio")  # Gi·ªçng Anh - Anh

        if not audio_tag:
            print(f"‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y √¢m thanh gi·ªçng Anh - Anh: {word}")
            return False

        # L·∫•y URL file MP3
        audio_url = audio_tag["data-src-mp3"]
        
        # T·∫£i file √¢m thanh
        audio_response = requests.get(audio_url, headers=HEADERS, timeout=10)
        if audio_response.status_code == 200:
            file_name = os.path.join(output_dir, f"{dem}.{word}.mp3")
            with open(file_name, "wb") as f:
                f.write(audio_response.content)
            print(f"‚úÖ ƒê√£ t·∫£i: {file_name}")
            return True
        else:
            print(f"‚ùå L·ªói t·∫£i file √¢m thanh: {word} (HTTP {audio_response.status_code})")
    
    except requests.exceptions.RequestException as e:
        print(f"‚ö†Ô∏è L·ªói k·∫øt n·ªëi khi t·∫£i {word}: {e}")

    return False
# L·∫∑p qua danh s√°ch t·ª´ v·ª±ng v√† t·∫£i file
def word_count(word):
    if(len(word.split(" "))>1):
        return "-".join(word.split())
    else: return word
dem = 1
for word in df["Word"]:
    word = word.strip().lower()  
    download_audio(word_count(word), dem)
    dem+=1
    #time.sleep(3)  # Ch·ªù 3 gi√¢y gi·ªØa m·ªói request ƒë·ªÉ tr√°nh b·ªã ch·∫∑n
    time.sleep(random.uniform(1, 3))
```

```python
import os
import time
import random
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

# ƒê·ªçc file t·ª´ CSV
file_path = r"E:\Project\2025_Project\English_audio_extractor_from_online_dict\unit2.csv"
df = pd.read_csv(file_path)

# T·∫°o th∆∞ m·ª•c l∆∞u file √¢m thanh
output_dir = r"E:\Project\2025_Project\English_audio_extractor_from_online_dict\audio"
os.makedirs(output_dir, exist_ok=True)

# Fake headers n√¢ng cao
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Referer": "https://www.collinsdictionary.com/",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
    "Cache-Control": "no-cache",
    "Pragma": "no-cache",
    "DNT": "1",
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1",
}

# Proxy (ƒëi·ªÅn th√¥ng tin proxy n·∫øu c·∫ßn)
PROXY = "username:password@proxy_ip:port"

# C·∫•u h√¨nh Selenium v·ªõi proxy
chrome_options = Options()
chrome_options.add_argument("--headless")  # Ch·∫°y ng·∫ßm
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1920x1080")
chrome_options.add_argument(f'--proxy-server=http://{PROXY}')  # Th√™m proxy n·∫øu c·∫ßn
chrome_options.add_argument(f'user-agent={HEADERS["User-Agent"]}')

# Kh·ªüi t·∫°o WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# H√†m x·ª≠ l√Ω t·ª´ th√†nh ƒë·ªãnh d·∫°ng URL h·ª£p l·ªá
def word_solve(word):
    return "-".join(word.split())

# H√†m t·∫£i file √¢m thanh b·∫±ng Selenium
def download_audio(word, dem):
    url = f"https://www.collinsdictionary.com/dictionary/english-word/{word}"

    try:
        driver.get(url)  # M·ªü trang web b·∫±ng Selenium
        time.sleep(random.uniform(2, 4))  # Ch·ªù t·∫£i trang

        # T√¨m th·∫ª ch·ª©a √¢m thanh
        audio_elements = driver.find_elements(By.CLASS_NAME, "hwd_sound sound audio_play_button icon-volume-up ptr")
        if not audio_elements:
            print(f"‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y √¢m thanh cho t·ª´: {word}")
            return False

        # L·∫•y URL file MP3
        audio_url = audio_elements[0].get_attribute("data-src-mp3")
        if not audio_url:
            print(f"‚ö†Ô∏è Kh√¥ng t√¨m th·∫•y URL MP3 cho t·ª´: {word}")
            return False

        # T·∫£i file √¢m thanh b·∫±ng requests (fake headers + proxy)
        session = requests.Session()
        session.headers.update(HEADERS)
        proxies = {
            "http": f"http://{PROXY}",
            "https": f"http://{PROXY}",
        }
        audio_response = session.get(audio_url, headers=HEADERS, proxies=proxies, timeout=10)
        
        if audio_response.status_code == 200:
            file_name = os.path.join(output_dir, f"{dem}.{word}.mp3")
            with open(file_name, "wb") as f:
                f.write(audio_response.content)
            print(f"‚úÖ ƒê√£ t·∫£i: {file_name}")
            return True
        else:
            print(f"‚ùå L·ªói t·∫£i file √¢m thanh: {word} (HTTP {audio_response.status_code})")
    
    except requests.exceptions.RequestException as e:
        print(f"‚ö†Ô∏è L·ªói k·∫øt n·ªëi khi t·∫£i {word}: {e}")

    return False

# L·∫∑p qua danh s√°ch t·ª´ v·ª±ng v√† t·∫£i file
dem = 1
for word in df["Word"]:
    word = word.strip().lower()  # Chu·∫©n h√≥a t·ª´
    word_url = word_solve(word)  # Chuy·ªÉn th√†nh d·∫°ng URL h·ª£p l·ªá
    print(f"üîé ƒêang x·ª≠ l√Ω: {word_url}")

    if download_audio(word_url, dem):
        dem += 1  # Ch·ªâ tƒÉng ƒë·∫øm n·∫øu t·∫£i th√†nh c√¥ng

    time.sleep(random.uniform(2, 5))  # Ch·ªù ng·∫´u nhi√™n gi·ªØa c√°c request

# ƒê√≥ng tr√¨nh duy·ªát Selenium
driver.quit()
```
