I. EXTRACT FROM OXFORD

```python
file_path = r"E:\Project\2025_Project\English_audio_extractor_from_online_dict\unit2.csv"
df = pd.read_csv(file_path)

output_dir = r"E:\Project\2025_Project\English_audio_extractor_from_online_dict\audio"
os.makedirs(output_dir, exist_ok=True)

# Thêm headers giả lập trình duyệt
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Referer": "https://www.oxfordlearnersdictionaries.com/",
    "Accept-Language": "en-US,en;q=0.9",
}

# Dùng Session để giữ kết nối
session = requests.Session()
session.headers.update(HEADERS)

def download_audio(word, dem):
    url = f"https://www.oxfordlearnersdictionaries.com/definition/english/{word}"
    
    try:
        response = requests.get(url, headers=HEADERS, timeout=10)  # Giới hạn timeout 10s
        if response.status_code != 200:
            print(f"❌ Không tìm thấy từ: {word} (HTTP {response.status_code})")
            return False

        # Phân tích HTML
        soup = BeautifulSoup(response.text, "html.parser")
        audio_tag = soup.find("div", class_="sound audio_play_button pron-uk icon-audio")  # Giọng Anh - Anh

        if not audio_tag:
            print(f"⚠️ Không tìm thấy âm thanh giọng Anh - Anh: {word}")
            return False

        # Lấy URL file MP3
        audio_url = audio_tag["data-src-mp3"]
        
        # Tải file âm thanh
        audio_response = requests.get(audio_url, headers=HEADERS, timeout=10)
        if audio_response.status_code == 200:
            file_name = os.path.join(output_dir, f"{dem}.{word}.mp3")
            with open(file_name, "wb") as f:
                f.write(audio_response.content)
            print(f"✅ Đã tải: {file_name}")
            return True
        else:
            print(f"❌ Lỗi tải file âm thanh: {word} (HTTP {audio_response.status_code})")
    
    except requests.exceptions.RequestException as e:
        print(f"⚠️ Lỗi kết nối khi tải {word}: {e}")

    return False
# Lặp qua danh sách từ vựng và tải file
def word_count(word):
    if(len(word.split(" "))>1):
        return "-".join(word.split())
    else: return word
dem = 1
for word in df["Word"]:
    word = word.strip().lower()  
    download_audio(word_count(word), dem)
    dem+=1
    #time.sleep(3)  # Chờ 3 giây giữa mỗi request để tránh bị chặn
    time.sleep(random.uniform(1, 3))
```

```python
import os
import time
import random
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

# Đọc file từ CSV
file_path = r"E:\Project\2025_Project\English_audio_extractor_from_online_dict\unit2.csv"
df = pd.read_csv(file_path)

# Tạo thư mục lưu file âm thanh
output_dir = r"E:\Project\2025_Project\English_audio_extractor_from_online_dict\audio"
os.makedirs(output_dir, exist_ok=True)

# Fake headers nâng cao
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Referer": "https://www.collinsdictionary.com/",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8",
    "Cache-Control": "no-cache",
    "Pragma": "no-cache",
    "DNT": "1",
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1",
}

# Proxy (điền thông tin proxy nếu cần)
PROXY = "username:password@proxy_ip:port"

# Cấu hình Selenium với proxy
chrome_options = Options()
chrome_options.add_argument("--headless")  # Chạy ngầm
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1920x1080")
chrome_options.add_argument(f'--proxy-server=http://{PROXY}')  # Thêm proxy nếu cần
chrome_options.add_argument(f'user-agent={HEADERS["User-Agent"]}')

# Khởi tạo WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# Hàm xử lý từ thành định dạng URL hợp lệ
def word_solve(word):
    return "-".join(word.split())

# Hàm tải file âm thanh bằng Selenium
def download_audio(word, dem):
    url = f"https://www.collinsdictionary.com/dictionary/english-word/{word}"

    try:
        driver.get(url)  # Mở trang web bằng Selenium
        time.sleep(random.uniform(2, 4))  # Chờ tải trang

        # Tìm thẻ chứa âm thanh
        audio_elements = driver.find_elements(By.CLASS_NAME, "hwd_sound sound audio_play_button icon-volume-up ptr")
        if not audio_elements:
            print(f"⚠️ Không tìm thấy âm thanh cho từ: {word}")
            return False

        # Lấy URL file MP3
        audio_url = audio_elements[0].get_attribute("data-src-mp3")
        if not audio_url:
            print(f"⚠️ Không tìm thấy URL MP3 cho từ: {word}")
            return False

        # Tải file âm thanh bằng requests (fake headers + proxy)
        session = requests.Session()
        session.headers.update(HEADERS)
        proxies = {
            "http": f"http://{PROXY}",
            "https": f"http://{PROXY}",
        }
        audio_response = session.get(audio_url, headers=HEADERS, proxies=proxies, timeout=10)
        
        if audio_response.status_code == 200:
            file_name = os.path.join(output_dir, f"{dem}.{word}.mp3")
            with open(file_name, "wb") as f:
                f.write(audio_response.content)
            print(f"✅ Đã tải: {file_name}")
            return True
        else:
            print(f"❌ Lỗi tải file âm thanh: {word} (HTTP {audio_response.status_code})")
    
    except requests.exceptions.RequestException as e:
        print(f"⚠️ Lỗi kết nối khi tải {word}: {e}")

    return False

# Lặp qua danh sách từ vựng và tải file
dem = 1
for word in df["Word"]:
    word = word.strip().lower()  # Chuẩn hóa từ
    word_url = word_solve(word)  # Chuyển thành dạng URL hợp lệ
    print(f"🔎 Đang xử lý: {word_url}")

    if download_audio(word_url, dem):
        dem += 1  # Chỉ tăng đếm nếu tải thành công

    time.sleep(random.uniform(2, 5))  # Chờ ngẫu nhiên giữa các request

# Đóng trình duyệt Selenium
driver.quit()
```
